```python
aisle_numbers_vertical_map = {}
for file in files:
  aisles_sorted_vertical = file.dropna().reset_index(drop=True)['raf'].apply(lambda x: x[:4]).unique().tolist()
  counter = 1
  side = 0
  for i, code in enumerate(aisles_sorted_vertical):
    aisle_numbers_vertical_map[code] = [40 - counter, side]
    if (i + 1) % 2 == 1:
      counter+=1
    side = (i + 1) % 2
```
